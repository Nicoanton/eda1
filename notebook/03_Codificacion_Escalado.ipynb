#### En esta sección se genera una división por el tipo de variable previamente establecido, categóricas y no categóricas (booleanas y continuas), para posteriormente generar un bucle que proporcione información acerca de las variables categóricas. 

```python
lista_col_cat = list(X_train.select_dtypes('category').columns)
lista_col_oth = list(set(X_train.columns).difference(lista_col_cat))
```

De igual manera existen variables nominales y ordinales, donde las nominales no tienen un orden específico entre sus categorías en comparación con las ordinales. En este caso a las variables nominales se les aplicará **One-Hot Encoding** y a las ordinales **Ordinal Encoding**.

#### Al visualizar la existencia de variables con elementos jerarquizables, se establece la siguiente lista de variables ordinales con sus respectivos ordenes, para posteriormente aplicar una función que genera la listas para cada tipo de codificación.

```python
ordinal_orders = { 'CNT_CHILDREN': list(range(0,20)),
                  'NAME_EDUCATION_TYPE': ['Lower secondary', 'Secondary / secondary special', 'Incomplete higher', 
                                          'Higher education', 'Academic degree'],
                  'REGION_RATING_CLIENT': [1,2,3],
                  'REGION_RATING_CLIENT_W_CITY': [1,2,3],
                  'NWEEKDAY_PROCESS_START': [1,2,3,4,5,6,7]
}
```

```python
ohe, ordinal, targ = f_aux.tipo_encoding(X_train, lista_col_cat, ordinal_orders)
```

### **Métodos de codificación**

```python
# One-Hot Encoding 
ohe = ce.OneHotEncoder(cols = ohe, use_cat_names = True) # Genera el nombre de la variable con la categoria
X_train_ohe = ohe.fit_transform(X_train)
X_test_ohe = ohe.transform(X_test)

# Ordinal Encoding
ordinal_encoder = OrdinalEncoder(categories = [ordinal_orders[col] for col in ordinal])
X_train_ohe[ordinal] = ordinal_encoder.fit_transform(X_train_ohe[ordinal])
X_test_ohe[ordinal] = ordinal_encoder.transform(X_test_ohe[ordinal])

ordinal_to_adjust = [col for col in ordinal if col != 'CNT_CHILDREN']
X_train_ohe[ordinal_to_adjust] = X_train_ohe[ordinal_to_adjust] + 1
X_test_ohe[ordinal_to_adjust] = X_test_ohe[ordinal_to_adjust] + 1

# Target Encoding
for col in targ:
    # Calculamos la media del target para cada categoría en los datos originales (s_train)
    # Usamos s_train, debido a que en esta dataframe esta contenida la variable objetivo 'TARGET'
    target_mean = s_train.groupby(col)['TARGET'].mean()
    
    # Aplicar la media calculada a las columnas de X_train_ohe y X_test_ohe
    X_train_ohe[col + '_encoded'] = X_train_ohe[col].map(target_mean)  # Crear nuevas columnas con '_encoded'
    X_test_ohe[col + '_encoded'] = X_test_ohe[col].map(target_mean)  # Aplica la misma transformación a X_test

    # Asegurando que las columnas codificadas sean float64
    X_train_ohe[col + '_encoded'] = X_train_ohe[col + '_encoded'].astype('float64')
    X_test_ohe[col + '_encoded'] = X_test_ohe[col + '_encoded'].astype('float64')

    # Eliminamos la columna anterior de donde se obtuvieron los datos
    if col in X_train_ohe.columns:
        X_train_ohe.drop(columns=[col], inplace=True)
    if col in X_test_ohe.columns:
        X_test_ohe.drop(columns=[col], inplace=True)

# Mostrar el resultado para verificar
print("Cantidad de columnas en X_train_ohe:", X_train_ohe.shape[1])
print("Cantidad de columnas en X_test_ohe:", X_test_ohe.shape[1])

# Ver las primeras filas de las columnas de X_train_ohe
print(X_train_ohe.head())
```

#### En la codificación de variables ordinales, se suma 1 a los resultantes de estas columnas tanto en el conjunto de entrenamiento como en el de prueba, para asegurar que los valores sigan el esquema establecido, donde los valores ordinales comienzan en 1 en lugar de 0. Este ajuste es necesario por que Python, por defecto, asigna índices que empiezan en 0, lo que desalinearía los datos con el esquema definido. Este ajuste no fue realizado en **CNT_CHILDREN**, debido a que al realizar la adición, ya no se representaría directamente la cantidad de hijos. 

#### Otro aspecto relevante a denotar en la codificación de las variables es que el *fit* solo se realiza en los datos de entrenamiento para que el codificador aprenda las categorías presentes en ellos. Luego, *transform* se usa en los datos de prueba para aplicar la misma codificación sin aprender nuevas categorías, asegurando que las columnas presenten la misma estructura en ambos conjuntos (entrenamiento y prueba).

```python
scaler = StandardScaler()
model_scaled = scaler.fit(X_train_ohe)
X_train_scaled = pd.DataFrame(scaler.transform(X_train_ohe), columns = X_train_ohe.columns, index = X_train_ohe.index)
X_test_scaled = pd.DataFrame(scaler.transform(X_test_ohe), columns = X_test_ohe.columns, index = X_test_ohe.index)
```
